```python
def subset_sum(arr, n, target):
    # Base cases
    if target == 0:
        return True
    if n == 0 or target < 0:
        return False

    # Include or exclude current element
    return (subset_sum(arr, n - 1, target - arr[n - 1]) or
            subset_sum(arr, n - 1, target))


# Input
n = int(input())
arr = list(map(int, input().split()))
target = int(input())

# Output
print(subset_sum(arr, n, target))
```

```python
def fibonacci(n, dp):
    if n <= 1:
        return n

    if dp[n] != -1:
        return dp[n]

    dp[n] = fibonacci(n - 1, dp) + fibonacci(n - 2, dp)
    return dp[n]


# Input
n = int(input())

# Initialize memoization array
dp = [-1] * (n + 1)

# Output
print(fibonacci(n, dp))
```
